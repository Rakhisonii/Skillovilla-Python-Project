```python
#python
import pandas as pd #importing pandas library

#TASK1 making the three dataframe
# making the employee data dictionary

employee_data={
    'ID':['A001','A002','A003','A004','A005'],                                                               #assigning the IDs to employee
    'Name':['John Alter', 'Alice Luxumberg', 'Tom Sabestine', 'Nina Adgra', 'Amy Johny'],                    #assigning names of employees
    'Gender':['M', 'F', 'M', 'F', 'F'],                                                                      #defining gender of employee
    'City':['Paris', 'London', 'Berlin', 'Newyork', 'Madrid'],                                               #defining city of employee
    'Age':[25,27,29,31,30]                                                                                   #defining age 
    }
employee_df = pd.DataFrame(employee_data)                                                  #changing dictionary to dataframe by using panda library
employee_df.to_csv('employee.csv', index=False)                                 #saving as csv file and setting index=false to not save index column 

employee_df                                                                     #preview of employee dataframe
```

```python
#Making the seniority level data dictionary

senioritylevel_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],                                                           #assigning IDs
    'Designation Level': [2, 2, 3, 2, 3]                                                                      #assigning designation level
}

senioritylevel_df = pd.DataFrame(senioritylevel_data)                                    #changing dictionary to dataframe by using panda library
senioritylevel_df.to_csv('senioritylevel.csv', index=False)                    #saving as csv file and setting index=false to not save index column 

senioritylevel_df                                                              #preview of senioritylevel dataframe
```

```python
#making project dictionary

project_data = {
    'ID': [
        'A001', 'A002', 'A003', 'A004', 'A005', 'A002', 'A005',                            #assigning IDs
        'A003', 'A001', 'A003', 'A001', 'A004', 'A004', 'A005'                             
    ],
    'Project': [ 
        'Project 1', 'Project 2', 'Project 3', 'Project 4', 'Project 5',                   #assigning projects
        'Project 6', 'Project 7', 'Project 8', 'Project 9', 'Project 10',
        'Project 11', 'Project 12', 'Project 13', 'Project 14'
    ],
    'Cost': [
        1002000, 2000000, 4500000, 5500000, None,                                          #assigning cost, the empty cost is giving as none
        680000, 400000, 350000, None, 300000,
        2000000, 1000000, 3000000, 200000
    ],
    'Status': [
        'Finished', 'Ongoing', 'Finished', 'Ongoing', 'Finished',                         #assigning status
        'Failed', 'Finished', 'Failed', 'Ongoing', 'Finished',
        'Failed', 'Ongoing', 'Finished', 'Finished'
    ]
}

project_df = pd.DataFrame(project_data)                                                   #changing dictionary to dataframe by using panda library
project_df.to_csv('project.csv', index=False)                                   #saving as csv file and setting index=false to not save index column 

project_df                                                                      #preview of project_df dataframe
```

```python
#TASK2: replacing missing values by running average

import pandas as pd
import numpy as np

# Step 1: Load saved project.csv
project_df = pd.read_csv('project.csv')

# Step 2: Running average fill using for loop
running_sum = 0
running_count = 0
filled_costs = []

for cost in project_df['Cost']:
    if pd.isna(cost):
        if running_count > 0:
            avg = running_sum / running_count
        else:
            avg = 0
        filled_costs.append(avg)
    else:
        running_sum += cost
        running_count += 1
        filled_costs.append(cost)

# Step 3: Replace old Cost column
project_df['Cost'] = filled_costs

# Step 4: Save updated project CSV
project_df.to_csv('project_filled.csv', index=False)

# Step 5: Preview
print(project_df)
```

```python
#TASK3: Spliting the name and removing older column

# Step 1: Load employee CSV
employee_df = pd.read_csv('employee.csv')

# Step 2: Split 'Name' column into 'First Name' and 'Last Name'
employee_df[['First Name', 'Last Name']] = employee_df['Name'].str.split(' ', expand=True)

# Step 3: Drop the original 'Name' column
employee_df.drop('Name', axis=1, inplace=True)

# Step 4: Save updated file (optional but recommended)
employee_df.to_csv('employee_updated.csv', index=False)

# Step 5: Preview the updated dataframe
print(employee_df)
```

```python
# Step 6: Reorder columns
employee_df = employee_df[['ID', 'First Name', 'Last Name', 'Gender', 'City', 'Age']]

# Step 7: Save updated CSV
employee_df.to_csv('employee_updated.csv', index=False)

#preview updates
print(employee_df)
```

```python
#Task4: joining all three dataframe in one dataframe named "Final"

#Step1: load all the CSVs needed to merge
employee_updated_df=pd.read_csv('employee_updated.csv')
senioritylevel_df=pd.read_csv('senioritylevel.csv')
project_filled_df=pd.read_csv('project_filled.csv')

#Step2: Merge employee table with seniority level table
merger=pd.merge(employee_updated_df,senioritylevel_df,on='ID')

#Step3: Merge the result with project table
Final=pd.merge(project_filled_df,merger,on='ID')

Final
```

```python
# Reorder columns after merge
Final = Final[['ID', 'First Name', 'Last Name', 'Gender', 'City', 'Age', 
               'Designation Level', 'Project', 'Cost', 'Status']]
#Step 5: Save to CSV
Final.to_csv('final_.csv', index=False)

#preview
Final
```

```python
#TASK5: adding new bonus column in Final dataframe

# Step 1: Add Bonus column (5% of cost if project is Finished)
Final['Bonus'] = Final.apply(
    lambda row: row['Cost'] * 0.05 if row['Status'].lower() == 'finished' else 0,
    axis=1
)

# Step 2: Save updated DataFrame
Final.to_csv('final_with_bonus.csv', index=False)

# Step 3: Preview
Final
```

```python
#task 6
import pandas as pd

# Step 1: Load the file (latest one before this task)
Final = pd.read_csv('final_with_bonus.csv')  # or use your current working version

# Step 2: Demote designation level by 1 where project failed
Final['Designation Level'] = Final.apply(
    lambda row: row['Designation Level'] + 1 if row['Status'].lower() == 'failed' else row['Designation Level'],
    axis=1
)

# Step 3: Remove rows where designation level > 4
Final = Final[Final['Designation Level'] <= 4]

# Step 4: Save the updated version
Final.to_csv('final_demoted.csv', index=False)

# Step 5: Preview
Final
```

```python
#TASK7
import pandas as pd

# Step 1: Load the last correct file (demoted one still has Gender)
Final = pd.read_csv('final_demoted.csv')

# Step 2: Add "Mr." or "Mrs." based on Gender
Final['First Name'] = Final.apply(
    lambda row: f"Mr. {row['First Name']}" if row['Gender'] == 'M' else f"Mrs. {row['First Name']}",
    axis=1
)

# Step 3: Drop the Gender column
Final.drop('Gender', axis=1, inplace=True)

# Step 4: Save updated DataFrame
Final.to_csv('final_titled.csv', index=False)

# Step 5: Preview
Final
```

```python
#TASK8:

import pandas as pd

# Step 1: Load the last saved file
Final = pd.read_csv('final_titled.csv')

# Step 2: Promote if age > 29 (but not below 1)
Final['Designation Level'] = Final.apply(
    lambda row: row['Designation Level'] - 1 if row['Age'] > 29 and row['Designation Level'] > 1 else row['Designation Level'],
    axis=1
)

# Step 3: Save updated version
Final.to_csv('final_promoted.csv', index=False)

# Step 4: Preview
Final
```

```python
#TASK9:

import pandas as pd

# Step 1: Load latest file (after promotions)
Final = pd.read_csv('final_promoted.csv')

# Step 2: Group by ID & First Name → Sum all project costs
TotalProjCost = Final.groupby(['ID', 'First Name'])['Cost'].sum().reset_index()

# Step 3: Rename column to 'Total Cost'
TotalProjCost.rename(columns={'Cost': 'Total Cost'}, inplace=True)

# Step 4: Save to new file
TotalProjCost.to_csv('total_project_cost.csv', index=False)

# Step 5: Preview
print(TotalProjCost)
```
